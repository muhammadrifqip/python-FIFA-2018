```python
import pandas as pd
import numpy as np
```

```python
data = pd.ExcelFile('data/FIFA World Cup 2018.xlsx')
```

```python
df_match = data.parse('match_statistics')
```

```python
df_countries = data.parse('countries')
```

```python
# df_match
```

#### Lowercase dulu 

```python
df_match['Team'] = df_match['Team'].str.lower()
df_match['Opponent'] = df_match['Opponent'].str.lower()
df_countries['Country'] = df_countries['Country'].str.lower()
df_countries['Continent'] = df_countries['Continent'].str.lower()
```

```python
df_countries.columns = ['Team','Continent_Team']
```

```python
df_match.merge(df_countries,on="Team",how='inner')
```

```python
df_match.groupby(['Team',"Goal Scored"],as_index=False).sum()
```

Get data with at least **Goal Scored** is 2 and **Passes Complete** more than 500

```python
df_match[(df_match['Goal Scored'] >= 2) & (df_match['Passes Completed'] > 500)]
```

Now, sort values based on **Goal Scored**, maximum to minimum

```python
df_match[(df_match['Goal Scored'] >= 2) & (df_match['Passes Completed'] > 500)].sort_values('Goal Scored', ascending = False)
```
